### Precipitation in the Boulder, CO Area

- According to the National Oceanic and Atmospheric Administration (NOAA), the average annual precipitation (including rain, snow, etc) in Boulder, CO is ~ 20 inches. 

- The majority of precipitation falls in the __Winter__ (snow) or in the __Summer__ (rain).

__Which months do you think have the highest precipitation, on average?__ Let's import some data to plot and see!

```python
# Numpy arrays can be created from text files
import numpy as np
# Import pyplot
import matplotlib.pyplot as plt

# import the monthly average values from `avg-monthly-precip.txt` as a numpy array
avg_monthly_precip = np.loadtxt(fname = "data/avg-monthly-precip.txt")
# print the dimensions of the data in the `avg_monthly_precip`
print(avg_monthly_precip.shape)

# import the names of the months from months.txt as a numpy array
months = np.genfromtxt("data/months.txt", dtype='str')
# print the dimensions of the data in the `months`
print(months.shape)
```

```python
################# FOR PLOTTING ####################
# Define plot space
fig, ax1 = plt.subplots(figsize=(10, 4))

# Define x and y axes
ax1.bar(months, 
        avg_monthly_precip,
        color = 'cyan')

# Set plot title and axes labels
ax1.set(title = "Average Monthly Precipitation in Boulder, CO",
       xlabel = "Month",
       ylabel = "Precipitation (inches/month)")

plt.show()
################# FOR PLOTTING ####################
```

__It looks like May and April get the most rainfall.__

In science, the standard convention is to work in the metric system. So we should probably convert. Let's convert to centimeters. There are 2.54 centimeters in 1 inch. Let's check the original values and then convert!

#### For these next questions, follow the instructions below:

- The csv file called `mo-precip-1971-2019-w-avg.csv` contains a list of monthly averages (12) per year (1971, 1972, etc), from 1971 - 2019. 
- The last line in the file contains the average monthly precipitation values for the _entire series,_ in our case, 1971 - 2019.

__Q3)__ Using the array code `.shape` (see example above), what are the dimensions of `mo-precip-1971-2019-w-avg.csv`? Keeping in mind those dimensions, how many years are actually represented? 

- _HINT: Should the last line count as an individual year?_


```python
################# QUESTION 3 ANSWER CODE ####################

# import the monthly average values from `monthly-precip-1971-2019.csv` as a numpy array
mo_precip_1971_2019 = np.loadtxt(fname= "data/mo-precip-1971-2019-w-avg.csv", delimiter = ",")

# print the dimensions of mo_precip_1971_2019
print(mo_precip_1971_2019.shape)
```

__Q3)__ __ANSWER__ 

> Dimensions are (50,12)

> 49 years are represented (1971 - 2019)

If you were to `print ()` the csv file `mo_precip_1971_2019`, you'd find there are a lot of rows! For this lesson, we __only__ want to work with the average of all the rows, i.e. _only the last line_. Since we are only interested in the last line, let's isolate it.

We can isolate the last line from the rest by creating a new `numpy array`

```python
################# SIMPLY RUN THIS CODE ####################

# select the first row and all twelve columns of monthly values
avg_mo_precip_1971_2019 = mo_precip_1971_2019[49:, 0:12]

# print data in `avg_mo_precip_1971_2019` to make sure something happened
avg_mo_precip_1971_2019
print(avg_mo_precip_1971_2019)
```

__Good!__
Now that we have extracted the average values for the climatology series, 1971 - 2019, let's get to work! 

__Q4)__ Convert the new climatology series from inches to centimeters (see example above). Use the `print()` to check.

```python
################# QUESTION 4 ANSWER CODE ####################

# Use assignment operator to convert array values from in to cm
avg_mo_precip_1971_2019 *= 2.54 # Unit Conversion: 2.54 cm = 1 in

# Print new array values to make sure the conversion was successful 
print(avg_mo_precip_1971_2019)
```

### Before plotting for QUESTION #5, ensure that `months` and `avg_mo_precip_1971_2019` have the SAME dimensions

Using the `.shape`, we see that `avg_mo_precip_1971_2019` is a two-dimensional array with a row count of 1 and a column count of 12. 

HOWEVER, `months` is a one-dimensional array. 

```python
################# SIMPLY RUN THIS CODE ####################

# print dimensions of `avg_mo_precip_1971_2019`
print(avg_mo_precip_1971_2019.shape)

# print dimensions of `months`
print(months.shape)
```

We cannot currently plot them as is. 

We must resize  the `avg_mo_precip_1971_2019` array to the correct dimensions:

__Q4.5)__ Use the `.resize` function to modify the array to match the months array __(12,)__

```python
################# QUESTION 4.5 ANSWER CODE ####################

# The .resize function modifies the array
# The .shape function tells you the dimensions in (XX,YY)

#Resize the array
avg_mo_precip_1971_2019.resize(12,)

# Print the values in the array to make sure the resizing went OK
print(avg_mo_precip_1971_2019)

# Print the shape of the array to make sure you now have the correct dimension (SEE EXAMPLE ABOVE)
print(avg_mo_precip_1971_2019.shape)

#Check that it matches the dimensions of the months array
print(months.shape)
```

__5)__ Copy the "FOR PLOTTING" block of code from above into your code and update the x and y axes, labels and titles with the correct terms and units. 

> Do the months of April, May and June still have the highest average precipitation? If not, what has changed?

```python
################# QUESTION 5 ANSWER CODE ####################

################# FOR PLOTTING ####################
# Define plot space
fig, ax2 = plt.subplots(figsize=(10, 4))

# Define x and y axes
ax2.bar(months, 
        avg_mo_precip_1971_2019, 
        color = 'green')
    

# Set plot title and axes labels
ax2.set(title = "Average Monthly Precipitation in Boulder, CO",
       xlabel = "Month",
       ylabel = "Precipitation (cm/month)")

plt.show()
################# FOR PLOTTING ###################
```

__BONUS KNOWLEDGE (the best, really):__ 

Let's do a plot v. plot comparison of the two climatologies. All you have to do here is run the code. 

```python
################# SIMPLY RUN THIS CODE ####################

# Define plot space
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

fig.suptitle("Average Monthly Precipitation, Boulder, CO", fontsize = 16)

# Define x and y axes
ax1.bar(months, 
       avg_monthly_precip,
       color = 'cyan', 
       edgecolor = 'darkblue')

# Set plot title and axes labels
ax1.set(title = "1971 - 2000 Climatology",
       xlabel = "Month",
       ylabel = "Precipitation\n(cm)");

# Define x and y axes
ax2.bar(months, 
        avg_mo_precip_1971_2019,
        color = 'green', 
        edgecolor = 'black')

# Set plot title and axes labels
ax2.set(title = "1971 - 2019 Climatology",
       xlabel = "Month",
       ylabel = "Precipitation\n(cm)")

plt.show()
```

#### Does anything seem different when comparing the two plots?

> The average precipitation from __Oct to Dec__ for both plots differs. In the __1971-2019__ plot, the precipitation seems to be __decreasing__ from October to December. That is certainly not the case in the 1971-2000 climatology.

> It seems then like the last 19 years have influenced the average monthly precipitation rates.
