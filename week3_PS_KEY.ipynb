### BACKGROUND

> When weather events occur and the resulting precipitation falls to the ground, a fraction of the water may become a part of __evaporation__ or can be absorbed by plant roots and returned to the atmosphere by __transpiration__, or, can become __runoff__. Water that does not become runoff or evapotranspiration, may infiltrate the ground surface. Ideally, this infiltrating water flux will flow downward through the __unsaturated zone__ until it reaches the water table __(figure below).__

> If a precipitation event is intense/extreme enough, the unsaturated zone may respond by pushing the infiltrating water towards the water table at a faster rate than usual, causing water table fluctuations to occur at faster rates than usual.

> Groundwater wells, which are located around the country, can sometimes have the right equipment to record water table fluctuations at varying depths in the subsurface and for certain time intervals, i.e. 10 minutes, 1 hours, etc. 

> Such records of water table measurements are a useful tool that scientists analyze to better understand how subsurface processes respond to precipitation events.

## What was the total rainfall from September 9th - September 16th?

### Let's use a `for` loop to find out!

> With a `for loop` we can "iterate", meaning do something repeatedly, over an iterable.

> An __iterable__ is an object that can return one of its elements at a time. Below are some examples of iterables:

> - Sequence types:  strings,  lists,  tuples
> - Non-sequence types:  dictionaries, files

> We are interested in knowing __how much precipitation fell during the event, from September 9th - 16th.__

> The list below contains a `numeric type` (ex: 6.4) and a `text type` (ex: 9th).

> We can use a `for` loop to sum the `numeric types` while leaving the `text types` alone. 

```python
####### No need to make changes to this, SIMPLY RUN THIS CODE #######

# List below includes the day and the precipitation amount (mm) 
# for the precipitation event that occurred in September 2013

precip = [("09th", 6.4), 
          ("10th", 25.9), 
          ("11th", 48.8), 
          ("12th", 230.6), 
          ("13th", 62), 
          ("14th", 0.3), 
          ("15th", 49.3), 
          ("16th", 12.4)]

# Starts For Loop
print("Summing the daily precipitation amounts during the event using a For Loop")

print(" ") #Spacing out the results
amount = 0
items = []
for date, mm in precip:
    print(" {} | ({})".format(date, mm))
    items.append(date)
    amount += mm
print(" ") #Spacing out the results

# print("\nEnd Date: {}".format(date))
print("Total Precipitation that fell from Sept. 9th - 16th: {}".format(amount), "mm")
# Ends For Loop
```
