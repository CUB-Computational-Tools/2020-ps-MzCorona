### BACKGROUND

> When precipitation events occur and fall to the ground, a fraction of the water may become a part of evaporation or can be absorbed by plant roots and returned to the atmosphere by transpiration, or, can runoff. Water that does not become runoff or evapotranspiration, may infiltrate the ground surface. In an ideal scenario, this infiltrating water flux will flow downward through the unsaturated zone until it reaches the water table __(see figure below).__

> If a precipitation event is intense/extreme enough, the unsaturated zone may respond by pushing the infiltrating water towards the water table at a faster rate than usual, causing water table fluctuations to occur.

> Groundwater wells are equipped to measure water table fluctuations, which helps scientists better understand how subsurface processes respond to precipitation events.
